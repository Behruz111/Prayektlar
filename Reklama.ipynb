```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
df = pd.read_csv('https://raw.githubusercontent.com/anvarnarz/praktikum_datasets/main/reklama_sales.csv')
df.head(20)
```

```python
plt.figure(figsize=(16,8))
sns.regplot(data=df,x=df['Sales'],y=df['TV'],line_kws={'color':'red'})
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.regplot(data=df,x='Sales',y='Radio',line_kws={'color':'red'})
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.regplot(data=df,x='Sales',y='Newspaper',line_kws={'color':'yellow'})
plt.show()
```

```python
from sklearn.model_selection import train_test_split
train_set,test_set=train_test_split(df,test_size=0.1,random_state=42)
```

```python
x=train_set['TV'].copy()
y=train_set['Sales'].copy()
```

```python
X=np.asanyarray(train_set[['TV']])
Y=np.asanyarray(train_set[['Sales']])
```

```python
type(X)
```

```python
X
```

```python
from sklearn import linear_model
LR_model=linear_model.LinearRegression()
```

```python
LR_model.fit(X,Y)
```

```python
x1=np.asanyarray(test_set[['TV']])
y1=np.asanyarray(test_set[['Sales']])
```

```python
LR_model.predict(x1)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
LMR=mean_absolute_error(y1,x1)
RMse=np.sqrt(mean_squared_error(y1,x1))
```
